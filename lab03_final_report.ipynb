**2. Write code to approximate the two signals using the Fourier series expansion above. Use 200 (instead of infinite number of) Sine waves. Then, plot these two signals.**

````python
# write your code here
N = 200

# Define the fundamental frequencies for both signals
f0_1 = 1 / 2  # For the first signal (T0 = 2)
f0_2 = 1 / 1  # For the second signal (T0 = 1)

# Time vector
t = np.linspace(-2, 2, 1000)  # Time interval for plotting

# Function to compute rect function (rectangular pulse)
def rect(t):
    return np.where(np.abs(t) <= 0.5, 1, 0)

# First signal s(t) = rect(t - 1/2)
def s1(t):
    return rect(t - 1/2)

# Second signal s(t) = rect(2t) - 1/2
def s2(t):
    return rect(2 * t) - 1/2

# Compute Fourier series approximation
def fourier_series_approx(s, f0, t, N):
    a0 = np.mean(s(t))  # The DC component
    approx = np.ones_like(t) * a0  # Start with the DC component

    for k in range(1, N + 1):
        # Fourier coefficients (can be computed analytically or numerically)
        Ak = 2 * np.mean(s(t) * np.sin(2 * np.pi * k * f0 * t))
        # Approximation sum
        approx += Ak * np.sin(2 * np.pi * k * f0 * t)
    
    return approx

# Compute the Fourier series approximation for both signals
s1_approx = fourier_series_approx(s1, f0_1, t, N)
s2_approx = fourier_series_approx(s2, f0_2, t, N)

# Plot the original signals and their Fourier approximations
plt.figure(figsize=(12, 6))

# Plot the first signal and its approximation
plt.subplot(2, 1, 1)
plt.plot(t, s1(t), label="Original Signal 1", color='b')
plt.plot(t, s1_approx, label="Fourier Approximation (N=200)", color='r', linestyle='--')
plt.title("Signal 1: rect(t - 1/2)")
plt.xlabel("Time [t]")
plt.ylabel("Amplitude")
plt.legend()

# Plot the second signal and its approximation
plt.subplot(2, 1, 2)
plt.plot(t, s2(t), label="Original Signal 2", color='b')
plt.plot(t, s2_approx, label="Fourier Approximation (N=200)", color='r', linestyle='--')
plt.title("Signal 2: rect(2t) - 1/2")
plt.xlabel("Time [t]")
plt.ylabel("Amplitude")
plt.legend()

plt.tight_layout()
plt.show()
````

```python
def DTFT(x,n0,w):
    """
    This function computes the DTFT of a discrete-time signal.
    
    Parameters
    ---
    x: the discrete-time signal
    n0: time index corresponding to the 1st element of the x vector
    w: frequencies
    
    Returns
    ---
    X: the computed DTFT
    """
    pass
```
**Note that if ```x``` is a vector of length $N$, then its DTFT is computed by**

\begin{equation}
    X(\omega)=\sum_{n=0}^{N-1}x[n]e^{-jw(n+n0)}
\end{equation}

**where $w$ is a vector that contains the frequencies from $-\pi$ to $\pi$.**

**Hint:** In Python, ```1j``` is defined as $\sqrt{-1}$. Use `np.exp(x)` to calculate $e^x$.

````python
# write your code here
def DTFT(x, n0, w):
    N = len(x)
    X = np.zeros(len(w), dtype=complex)
    
    for i, omega in enumerate(w):
        summation = np.sum(x * np.exp(-1j * omega * (np.arange(N) + n0)))
        X[i] = summation
        
    return X
````

**2. For the following signals** 

* $x[n]=\delta[n]$
* $x[n]=\delta[n-5]$
* $x[n]=(0.5)^nu[n]$

**use your DTFT function to compute $X(\omega)$, and plots its magnitude and phase.**

**Hint**: Use `np.power(a,b)` to calculate $a^b$. Use ```np.abs()``` and ```np.angle()``` to compute the magnitude and phase.

````python
def delta(n):
    return np.where(n == 0, 1.0, 0.0)

def u(n):
    return np.where(n >= 0, 1.0, 0.0)

N = 20
n = np.arange(N)
w = np.linspace(-np.pi, np.pi, 500)
x1 = delta(n)
x2 = delta(n - 5)
x3 = np.power(0.5, n) * u(n)
X1 = DTFT(x1, 0, w)
X2 = DTFT(x2, 0, w)
X3 = DTFT(x3, 0, w)
signals = [(X1, 'x[n] = delta[n]'), (X2, 'x[n] = delta[n-5]'), (X3, 'x[n] = (0.5)^n u[n]')]

for X, title in signals:
    magnitude = np.abs(X)
    phase = np.angle(X)

    plt.figure()
    plt.subplot(2, 1, 1)
    plt.plot(w, magnitude)
    plt.title(f'Magnitude of {title}')
    plt.xlabel('Frequency (rad/sample)')
    plt.ylabel('|X(w)|')

    plt.subplot(2, 1, 2)
    plt.plot(w, phase)
    plt.title(f'Phase of {title}')
    plt.xlabel('Frequency (rad/sample)')
    plt.ylabel('Phase of X(w) (radians)')
    
    plt.tight_layout()
    plt.show()
````
